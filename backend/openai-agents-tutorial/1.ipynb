```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()

model = OpenAI(
    api_key=os.getenv('API_KEY'),
    base_url=os.getenv('BASE_URL')
)


result = model.responses.create(
    model=os.getenv('MODEL'),
    input='AI大模型领域最新的顶会论文',
    tools=[{'type':'web_search'}]
)

print(result.output_text)
```

```python
import os
from openai import AsyncOpenAI
from dotenv import load_dotenv
from agents import (
    Agent, 
    Runner,
    set_default_openai_client,
    set_tracing_disabled,
)
set_tracing_disabled(True)
load_dotenv()


# 1.创建异步OpenAI客户端
client = AsyncOpenAI(
    api_key=os.getenv("API_KEY"), 
    base_url=os.getenv("BASE_URL")
)

# 2.全局注册客户端，所有的Agent实例都会使用这个客户端
set_default_openai_client(client)

async def main():
    # 3.创建一个Agent实例
    agent = Agent(
        name="文学专家",
        instructions='你只会用七言绝句回答',
        model=os.getenv("MODEL"),
    )

    result = await Runner.run(agent, "写一首春天的七言绝句")
    print(result.final_output)


await main()
```

```python
from agents import OpenAIResponsesModel, Model, ModelProvider, RunConfig

# 1.自定义ModelProvider类，实现ModelProvider接口的get_model方法
class MyModelProvider(ModelProvider):

    def get_model(self, model_name: str)-> Model:
        print(model_name)
        # 创建一个OpenAIResponsesModel对象，并返回
        return OpenAIResponsesModel(model=model_name, openai_client=client)

MyModelProvider = MyModelProvider()

async def main():
    # 使用自定义的MyModelProvider对象创建一个Agent对象
    agent = Agent(name='Assitant', instructions='你只会用七言绝句回答', model=os.getenv('MODEL'))

    result = await Runner.run(agent, input='写一首春天的七言绝句', run_config=RunConfig(model_provider=MyModelProvider))

    print(result.final_output)
    
await main()
```
